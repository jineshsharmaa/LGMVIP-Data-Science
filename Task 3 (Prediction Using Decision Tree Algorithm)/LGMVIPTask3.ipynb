```python
import pandas as pd
data = load_iris()
dataframe = pd.DataFrame(data.data, columns = data.feature_names)
dataframe['target'] = data.target
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels'] 
dataframe.head(10)
```
